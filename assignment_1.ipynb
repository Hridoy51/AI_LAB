# Import the libraries

```python
import numpy as np
import matplotlib.pyplot as plt
```

# Define and Initializing X

```python
x = np.arange(-10,10.1,0.1)
```

# plot y = wx + b

```python
w,b = .3,.5
y = w*x + b
plt.plot(x,y)
plt.title("y = wx + b")
plt.xlabel("X values")
plt.ylabel("Y values")
```

#  plot y = x^2

```python
y = x*x
plt.plot(x,y)
plt.title("y = x^2")
plt.xlabel("X values")
plt.ylabel("Y values")
```

# plot y = 1/ (1 + e^(-x))

```python
y = 1/(1+np.exp(-x))
plt.plot(x,y)
plt.title("y = 1/ (1 + e^(-x))")
plt.xlabel("X values")
plt.ylabel("Y values")
```

# Plot y = (e^x - e^(-x)) / (e^x - e^(-x))

```python
y = (np.exp(x) - np.exp(-x)) / (np.exp(x) + np.exp(-x))
plt.plot(x,y)
plt.title("y = (e^x - e^(-x)) / (e^x + e^(-x))")
plt.xlabel("X values")
plt.ylabel("Y values")
```

# Plot y = g(f(x)) where u = f(x) = wx + b and y = g(u) = 1/ (1 + e^(-u))

```python
w,b = 1,.5
u = w*x + b
y = 1/(1+np.exp(-u))
plt.plot(x,y)
plt.xlabel("X values")
plt.ylabel("Y values")
```

# Plot y = g(f(x)) where f(x) = wx + b, u = f(x), and g(u) = (e^u - e^(-u)) / (e^u + e^(-u))

```python
w,b = .5,.3
u = w*x + b
y = (np.exp(u)-np.exp(-u))/(np.exp(u)+np.exp(-u))
plt.plot(x,y)
plt.xlabel("X values")
plt.ylabel("Y values")
```

# Ploting a complex function where first clculate y1 and y2 then y and plot the y, here are the functions

# y1 = g1(f1(x)) where f1(x) = w1x + b1, u1 = f1(x), and g1(u1) = 1/ (1 + e^(-u1))
# y2 = g2(f2(x)) where f2(x) = w2x + b2, u2 = f2(x), and g2(u2) = 1/ (1 + e^(-u2))
# y = g3(f3(v)) where f3(v) = w3y1 + w4y2 + b, w = f3(v), and g3(w) = 1/ (1 + e^(-w))

```python
# calculate y1
w1,b1 = .5,.3
u1 = w1*x + b1
y1 = 1 / (1 + np.exp(-u1))

#calculate y2
w2,b2 = 1,.75
u2 = w2*x + b2
y2 = 1 / (1 + np.exp(-u2))

# clculate Y
w3,w4,b = .33,.44,2
w = w3*y1 + w4*y2 + b
y = 1/ (1+np.exp(-w))

plt.plot(x,y)
plt.xlabel("X values")
plt.ylabel("Y values")
```
