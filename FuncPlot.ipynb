### Imports

```python
import matplotlib.pyplot as plt
import numpy as np
```

### Define a, b and x

```python
a,b = 3,5
x = np.array(range(-5,6))
```

### Ploting y = ax + b

```python
y = a*x + b
plt.plot(x,y, linestyle='--', marker='o', color='b', label='line with marker')
for i in range(len(x)):
    plt.annotate(f"({x[i]},{y[i]})", (x[i],y[i]))
plt.title("Y = aX + b")
plt.legend()
```

### Ploting y = asin(x)+b

```python
y = a*np.sin(x) + b
plt.plot(x,y, linestyle='--', marker='o', color='b', label='line with marker')
for i in range(len(x)):
    plt.annotate(f"({np.round(x[i],2)},{np.round(y[i],2)})", (x[i],y[i]))
plt.title("Y = asin(x) + b")
plt.legend()
```

### Ploting y = ax^2 + b

```python
y = a*x*x + b
plt.plot(x,y, linestyle='--', marker='o', color='b', label='line with marker')
for i in range(len(x)):
    plt.annotate(f"({np.round(x[i],2)},{np.round(y[i],2)})", (x[i],y[i]))
plt.title("Y = ax^2 + b")
plt.legend()
```

### Ploting y = ae^x + b

```python
y = a*np.exp(x) + b
plt.plot(x,y, linestyle='--', marker='o', color='b', label='line with marker')
for i in range(len(x)):
    plt.annotate(f"({np.round(x[i],2)},{np.round(y[i],2)})", (x[i],y[i]))
plt.title("Y = ae^x + b")
plt.legend()
```
